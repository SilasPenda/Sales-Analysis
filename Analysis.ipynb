##### Convert columns to Integers

```python
# Make 'Quantity' an Integer
all_data['Quantity'] = pd.to_numeric(all_data['Quantity'])
# Make 'Price Each' a Float
all_data['UnitPrice'] = pd.to_numeric(all_data['UnitPrice'])
```

##### Add relevant columns

```python
# Convert 'InvoiceDate' column to datetime format
all_data['InvoiceDate']=pd.to_datetime(all_data['InvoiceDate'])
all_data.head()
```

```python
# Add Hour, Month, & Year columns
all_data['Hour'] = pd.to_datetime(all_data['InvoiceDate'], format= '%Y-%m-%d %H:%M:%S').dt.hour
all_data['Month'] = pd.to_datetime(all_data['InvoiceDate'], format= '%Y-%m-%d %H:%M:%S').dt.month_name()
all_data['Year'] = pd.to_datetime(all_data['InvoiceDate'], format= '%Y-%m-%d %H:%M:%S').dt.year
all_data.head()
```

```python
# Add Sales column
all_data['Sales'] = all_data['Quantity'] * all_data['UnitPrice']
all_data.head()
```

### Question 1: What is the best sales month and how much was earned that month?

```python
all_data['Month'] = pd.Categorical(all_data['Month'], ordered=True, categories= ['January', 'February', 'March', 'April','May','June', 'July', 'August','September', 'October', 'November', 'December'])
all_data.pivot_table(index='Month', values='Sales', aggfunc='sum')
```

```python
group = all_data.groupby('Month').sum()

plt.figure(figsize=(8,5))
Month = [Month for Month, all_data in all_data.groupby('Month')] # This arranges the months in the right order
plt.title('Best Sales Month', fontdict={'fontweight':'bold', 'fontsize':18})
plt.bar(Month, group['Sales'])
plt.xticks(Month, rotation='vertical', size=10)
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.ticklabel_format(useOffset=False, style='plain', axis='y') #Make value on y axis to be plain and not exponential

plt.show()
```

### Question 4: What products are most often sold together?

```python
all_data.rename(columns={'Description':'Product'}, inplace=True)
all_data.head()
```

```python
df = all_data[all_data['InvoiceNo'].duplicated(keep=False)] # This creates a dataframe and stores all duplicated 'InvoiceNo' and keeps them all

# Create a new column called 'Grouped' to store the products sold together
df['Grouped'] = df.groupby('InvoiceNo')['Product'].transform(lambda x: ','.join(x))
# Drop duplicate 'Order ID'
df = df[['InvoiceNo', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2))) # This counts the 2 products most often sold together.
for key, value in count.most_common(10):
    print(key, value)
```

### Question 5:  What product sold the most?

```python
# Plotting the products sold the most on a chart will not be visible enough because there are too many producst sold.

# Create a pivot table for the sum of quntity of products ordered and sort them by 'Quantity' in descending order
quantity_ordered = all_data.pivot_table(index='Product', values='Quantity', aggfunc='sum')
top_products = quantity_ordered.sort_values(by='Quantity', ascending=False) # Sort 'Products' in descending order by 'Quantity'
top_products.head()
```

```python
# Store sorted products in dataframe
quantitydf = top_products.reset_index()
# Store top 20 products in dataframe
quantitydf_20 = quantitydf.sort_values(by='Quantity', ascending=False).iloc[0:20]
quantitydf_20
```

```python
plt.bar(quantitydf_20['Product'], quantitydf_20['Quantity'])
plt.xticks(quantitydf_20['Product'], rotation='vertical', size=8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product Name')
plt.ticklabel_format(useOffset=False, style='plain', axis='y') #Make value on y axis to be plain and not exponential
plt.show()
```
